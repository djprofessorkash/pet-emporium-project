## Exploratory Data Analysis

```python
import os

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
DATAPATH = "../data/images/"
```

```python
clean_direcs = []
for subdirec, direcs, files in os.walk(DATAPATH):
    for direc in direcs:
        if "n02" in direc:
            clean_direc = " ".join(token.capitalize() for token in direc.split("-")[1].split("_"))
            clean_direcs.append(clean_direc)
```
